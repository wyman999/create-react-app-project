## 3. Launch Application and Generate Public URL

Now we'll start the Streamlit server and create an ngrok tunnel to make it accessible online:

```python
import subprocess
import time
import os

print("🚀 Starting MoneyPrinterTurbo...")

# 设置环境变量，让 Streamlit 允许外部访问
os.environ['STREAMLIT_SERVER_HEADLESS'] = 'true'

# 启动 Streamlit 服务器
streamlit_proc = subprocess.Popen([
    "uv", "run", "streamlit", "run", "./webui/Main.py",
    "--server.port=8501",
    "--server.address=0.0.0.0",
    "--server.headless=true"
])

print("✅ MoneyPrinterTurbo is starting...")
print("⏳ Please wait 10-20 seconds, then check the link below!")
print("\n" + "="*60)
print("🌐 Access URL: https://localhost:8501")
print("   (Colab will automatically create a public link for you)")
print("="*60)
```
